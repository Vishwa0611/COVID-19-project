```python
vaccine_df=pd.read_csv("D:/COVID-19 project/archive/covid_vaccine_statewise.csv")
```

```python
# droping unnecessary columns
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"], inplace= True, axis = 1)
```

```python
covid_df.head()
```

```python
# changed format of date 
covid_df['Date']= pd.to_datetime(covid_df['Date'],format =  '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
# active cases
covid_df['Active_cases']= covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
# pivot table
statewise= pd.pivot_table(covid_df, values= [ "Confirmed", "Deaths", "Cured"],
                          index = "State/UnionTerritory", aggfunc=max)
```

```python
statewise["Recovery Rate"]= statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"]= statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise= statewise.sort_values(by = "Confirmed", ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
# top 10 active 
top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by='Active_cases', ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(16, 9))
```

```python
plt.title("Top 10 state with most active cases in INDIA", size = 25)
```

```python
axis= sns.barplot(data =top_10_active_cases.iloc[:10], y= "Active_cases", x= "State/UnionTerritory", linewidth= 2, edgecolor= 'red')
```

```python
# top 10 active 
top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by='Active_cases', ascending=False).reset_index()
fig = plt.figure(figsize=(16, 9))
axis= sns.barplot(data =top_10_active_cases.iloc[:10], y= "Active_cases", x= "State/UnionTerritory", linewidth= 2, edgecolor= 'red')
plt.title("Top 10 state with most active cases in INDIA", size = 25)
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# Top states with highest death
top_10_deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by='Deaths', ascending=False).reset_index()

fig= plt.figure(figsize=(18,4))

plt.title("Top 10 states with most Deaths", size= 20)
ax= sns.barplot(data= top_10_deaths.iloc[:12],y = "Deaths", x="State/UnionTerritory", linewidth=3, edgecolor= 'Green')
plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
# Growth trend
fig= plt.figure(figsize=(10,6))
ax= sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x= 'Date', y='Active_cases', hue= 'State/UnionTerritory')
ax.set_title("Top 5 Affected States in INDIA", size = 16)
```

```python
# remove rows (state=INDIA)
vaccine= vaccine_df[vaccine_df.State !='India']
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated": "Total"}, inplace=True)
```

```python
vaccine
```

```python
# most vaccinated state
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac= max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python

fig = plt.figure(figsize=(12, 5))
x= sns.barplot(data =max_vac.iloc[:10], y= max_vac.Total, x= max_vac.index, linewidth= 2, edgecolor= 'red')
plt.title("Top 5 vaccinated state in INDIA", size = 25)
plt.xlabel("States")
plt.ylabel("Vaccinations")
plt.show()
```

```python
fig = plt.figure(figsize=(12, 5))
x = sns.barplot(
    data=max_vac.iloc[:10],
    y="Total",
    x=max_vac.index,
    hue=max_vac.index,           # Assign hue
    legend=False,                # Hide legend
    linewidth=2,
    edgecolor='black',
    palette='husl'
)
plt.title("Top 5 Vaccinated States in INDIA", size=25)
plt.xlabel("States")
plt.ylabel("Vaccinations")

plt.tight_layout()
plt.show()
```

```python
# least vaccinated states
min_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac = min_vac.sort_values('Total', ascending=True).head(5)
min_vac
```
